## Tarefas:

# 1 - Registro de Empréstimos:

# Registre os empréstimos de livros para os  alunos fictícios da escola.
# 2 -  Avaliação de Livros:

# Após os empréstimos, simule que os alunos leram os livros e atribua notas de avaliação aos livros emprestados.

# 3 - Avaliação do Atendimento da Biblioteca:

#  Registre avaliações do atendimento da biblioteca por parte dos alunos.

# 4 - Imprimir Relatório:

# Após realizar as ações anteriores, imprima um relatório que apresente as informações sobre os empréstimos, avaliações dos livros e a média das avaliações do atendimento da biblioteca.


```python
# Dicionário para armazenar os empréstimos
emprestimos = []

# Empréstimo 1
aluno1 = input("Digite o nome do primeiro aluno: ")
livro1 = input("Digite o título do livro emprestado: ")
data_emprestimo1 = input("Digite a data do empréstimo (dd/mm/aaaa): ")

emprestimo1 = {
    'aluno': aluno1,
    'livro': livro1,
    'data_emprestimo': data_emprestimo1,
}

emprestimos.append(emprestimo1)

# Empréstimo 2
aluno2 = input("Digite o nome do segundo aluno: ")
livro2 = input("Digite o título do livro emprestado: ")
data_emprestimo2 = input("Digite a data do empréstimo (opcional): ")

emprestimo2 = {
    'aluno': aluno2,
    'livro': livro2,
    'data_emprestimo': data_emprestimo2,
}

emprestimos.append(emprestimo2)

# Dicionário para armazenar as avaliações
avaliacoes = {
    livro1: int(input(f"Digite a nota de avaliação para '{livro1}': ")),
    livro2: int(input(f"Digite a nota de avaliação para '{livro2}': ")),
    # Adicione mais avaliações aqui
}

# Lista para armazenar as avaliações do atendimento
avaliacoes_atendimento = []

# Coleta das avaliações
for _ in range(2):  # Vamos coletar 2 avaliações
    nota = int(input("Digite a nota de atendimento (de 1 a 5): "))
    avaliacoes_atendimento.append(nota)

# Calcula a média das avaliações do atendimento
media_atendimento = sum(avaliacoes_atendimento) / len(avaliacoes_atendimento)
print(f"Média das avaliações do atendimento: {media_atendimento:.2f}")
```
